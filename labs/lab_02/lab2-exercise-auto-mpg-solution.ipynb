### Missing Data

```python
# Horsepower has missing data (marked as ?).  
# Can you set the value to Average horse power from other cars that have same number of Cylinders?

# First we need to convert the horsepower to a numeric field before we can take the mean (average) values
# To do this, we need to replace ? with a valid numeric field - our choices - 0 and NaN.
# If we set them to 0, it will affect the mean values.  
# But when we set it to NaN (Not a Number), those rows are skipped when computing mean.
auto_mpg.horsepower.unique()
# Notice the question mark in the third row.
```

```python
# Replace the ? with NaN value
auto_mpg.horsepower = auto_mpg.horsepower.replace('?', np.nan)
auto_mpg.info()
```

```python
# Horsepower is marked as "object" datatype because of missing data - check using info()
# Now that you have filled the missing data, can you change the datatype to integer? - np.int64?

# Set the type of horsepower to Float
auto_mpg.horsepower = auto_mpg.horsepower.astype(float)
auto_mpg.horsepower.mean()
```

### Challenge - Feature Extraction

```python
# Can you separate the Car Company from car_name and count how many cars are in the dataset by car company?
# Assume the first word in car_name is the car company
auto_mpg.car_name[0]
```

```python
# Lets split this value and get the first part of the string
auto_mpg.car_name[0].split()[0]
```

```python
# Now we have the logic, let's apply this to all the values in the DF
def get_car_company(car_name):
    return car_name.split()[0]

auto_mpg['car_company'] = auto_mpg.car_name.map(get_car_company)
```

```python
# Notice that we now have Car Company as a column
auto_mpg.head(5)
```

```python
# How many 4-6-8 cylinder cars are there in the dataset for each car company
# Pivot Table makes this easy
pd.pivot_table(auto_mpg, index='car_company', columns='cylinders', values='mpg', aggfunc=np.size).fillna(0)
```

```python
# Lets verify this via Volvo
# Notice that it has 4 rows with Cylinder-4 and 2 with Cylinder-6
auto_mpg[auto_mpg['car_company'] == 'volvo']
```
